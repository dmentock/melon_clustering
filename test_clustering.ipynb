```python
from melon_clustering import PatternExtractor
# Initialize and use PatternExtractor
extractor = PatternExtractor('jp', radius=5, filter_threshold=3)
target_char = '悦'
# Example usage:
sentences_dict = {'悦': [
    '⦅弱ぇ者 殺して悦に浸る➡',
    '悦に浸りたいだけなんだよ！',
    '俺は君を利用して悦に浸ってた',
    '悦に入っているだけの傲慢な男に',
    '悦に浸る鴉を遠く眺めていたんだ',
    'シンプルな答えを 複雑にして悦に浸るの',
]}
# Step 1: Extract patterns
sentence_patterns_flat = extractor.extract_patterns_flat(sentences_dict[target_char], target_char)

# Step 2: Filter only patterns containing <MASK>
extractor.filter_mask_patterns(sentence_patterns_flat)

# Step 3: Map sentences to pattern indices
sentence_pattern_mapping = extractor.map_sentences_to_patterns(sentences_dict[target_char], target_char)

# Step 4: Plot the PCoA results
extractor.plot_pcoa(sentence_pattern_mapping)
```
