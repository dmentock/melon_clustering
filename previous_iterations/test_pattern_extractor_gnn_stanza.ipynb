In this notebook, we convert sentences into their abstract stanza pos/deprel representation first before generating the clusters

```python
import pandas as pd
from melon_clustering import PatternExtractorGNN, ClusterEvaluator, CACHE_DIR, Loader, reference_clusters_de
import re

dim_reduction_methods = ['LSA', 'PCA', 't-SNE']
clustering_methods = ['KMeans', 'Agglomerative', 'DBSCAN']
results_collection = []


def extract_tags(sentences, get_pos=True, get_deprel=True):
    converted_sentences = []
    for sentence_info in sentences:
        converted_sentence = []
        sentence = sentence_info['sentence']
        pattern = r'<(\d+):([^/]+)/([^>]+)>\(([^)]+)\)'
        parts = sentence.split('<ROOT>')
        for part_index, part in enumerate(parts):
            for match in re.finditer(pattern, part):
                word_id, pos, deprel, word = match.groups()
                tag = f'{pos if get_pos else ""}{"/" if get_deprel and get_pos else ""}{deprel if get_deprel else ""}'
                converted_sentence.append(tag)
            if part_index < len(parts) - 1:
                converted_sentence.append('<ROOT>')
        converted_sentences.append(' '.join(converted_sentence))
    return converted_sentences

# for sigmoid_steepness in [0.2, 0.4, 0.6, 0.8, 1]:
#     for overlap_threshold in [0.2, 0.4, 0.6, 0.8, 1]:
#         for n_sentences in [5, 50, 500, 5000, 10000]:
import time
start = time.time()

for stanza_tags in [(True,True),(True,False),(False,True)]:

    stanza_pos = stanza_tags[0]
    stanza_deprel = stanza_tags[1]
    reference_clusters_preprocessed_by_word = {}
    for ref_word, reference_clusters in reference_clusters_de.items():
        reference_clusters_preprocessed = []
        for cluster in reference_clusters:
            stanza_sentences = Loader.preprocess_stanza([sentence[0].replace(sentence[1], '<ROOT>') for sentence in cluster], 'de')
            processed_sentences = extract_tags(stanza_sentences, get_pos=stanza_pos, get_deprel=stanza_deprel)
            processed_cluster = []
            for i in range(len(processed_sentences)):
                processed_cluster.append([processed_sentences[i], cluster[i][1]])
            reference_clusters_preprocessed.append(processed_cluster)
        reference_clusters_preprocessed_by_word[ref_word] = reference_clusters_preprocessed


    for sigmoid_steepness in [0.2]:
        for overlap_threshold in [0.2]:
            for n_sentences in [5]:
                for ref_word, reference_clusters in reference_clusters_de.items():
                    print("processing", sigmoid_steepness, overlap_threshold, n_sentences, ref_word, time.time() - start)
                    start = time.time()
                    additional_sentences_dict = Loader.load_sentences_from_word(ref_word, 'de', n_sentences=n_sentences)
                    additional_sentences_dict_stanza = Loader.load_sentences_from_word(ref_word, 'de', n_sentences=n_sentences, preprocessor='stanza')

                    # evaluator = ClusterEvaluator(reference_clusters=reference_clusters)
                    # evaluator.add_reference_sentences(reference_clusters)
                    # evaluator.add_additional_sentences(additional_sentences_dict)
                    # evaluator_preprocessed = evaluator
                    # print("reference_clusters",reference_clusters)
                    # print("additional_sentences_dict",additional_sentences_dict)

                    additional_sentences_dict_preprocessed = {}
                    for morphology, stanza_sentences in additional_sentences_dict_stanza.items():
                        additional_sentences_dict_preprocessed[morphology] = extract_tags(stanza_sentences, get_pos=stanza_pos, get_deprel=stanza_deprel)

                    reference_clusters_preprocessed = reference_clusters_preprocessed_by_word[ref_word]
                    # print("reference_clusters_preprocessed",reference_clusters_preprocessed)
                    # print("additional_sentences_dict_preprocessed",additional_sentences_dict_preprocessed)
                    evaluator_preprocessed = ClusterEvaluator(reference_clusters=reference_clusters_preprocessed)
                    evaluator_preprocessed.add_reference_sentences(reference_clusters_preprocessed)
                    evaluator_preprocessed.add_additional_sentences(additional_sentences_dict_preprocessed)
                    # for i,y  in evaluator.sentences_dict.items():
                    #     print("ee", i, y)
                    # for i,y  in evaluator_preprocessed.sentences_dict.items():
                    #     print("ep", i, y)

                    extractor = PatternExtractorGNN()
                    extractor.initialize(evaluator_preprocessed.sentences_dict, overlap_threshold=overlap_threshold)
                    extractor.initialize_node_embeddings()
                    embeddings, _ = extractor.get_all_sentence_embeddings(evaluator_preprocessed.sentences_dict, steepness=sigmoid_steepness)

                    results_df = evaluator_preprocessed.run_all_configurations(
                        vectors=embeddings,
                        ref_word=ref_word,
                        overlap_threshold=overlap_threshold,
                        n_sentences=n_sentences,
                        sigmoid_steepness=sigmoid_steepness,
                        dim_methods=dim_reduction_methods,
                        cluster_methods=clustering_methods,
                        sentence_paths=[(sid, evaluator_preprocessed.sentence_id_to_original[sid]) for sid in evaluator_preprocessed.reference_sentence_ids],
                        plot_seaborn=True,
                        plot_plt=False,
                        annotate_plt=False,
                        save=False,
                        stanza_pos = stanza_pos,
                        stanza_deprel = stanza_deprel,
                    )

                    results_collection.append(results_df)

# Combine all results using pd.concat
df = pd.concat(results_collection, ignore_index=True)
df.to_csv(CACHE_DIR / 'final_clustering_results_gnn.csv', index=False)
df
```
